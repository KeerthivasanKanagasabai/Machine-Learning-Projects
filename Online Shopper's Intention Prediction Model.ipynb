```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
df= pd.read_csv('online_shoppers_intention.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder

enc = LabelEncoder()
df.Month = enc.fit_transform(df.Month)
df.VisitorType = enc.fit_transform(df.VisitorType)
df.Weekend  =enc.fit_transform(df.Weekend)
df.Revenue = enc.fit_transform(df.Revenue)
```

```python
df
```

```python
df.dtypes
```

```python
x = df.drop(['Revenue'], axis = 1)
y = df.Revenue

x
```

```python
y
```

```python
x.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split 

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2)
```

```python
print('Training Independent Variable dataset Shape: ',x_train.shape)
print('Training dependent Variable dataset Shape: ',y_train.shape)
print('Test Independent Variable dataset Shape: ',x_test.shape)
print('Test dependent Variable dataset Shape: ',y_test.shape)
```

### Model 1 -- Extra Tree Classifier

```python
from sklearn.ensemble import ExtraTreesClassifier

model = ExtraTreesClassifier(n_estimators = 100, criterion = 'gini',max_depth = 500, max_features = 'log2')
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

accuracy = accuracy_score(y_test, y_pred)
accuracy

print(f" ML using Extra Trees Classifier yields accuracy of {accuracy * 100} %")
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
sns.heatmap(data = cm, annot = True)
```

### Model 2 -- Neural Network Model (Multi Layer Preceptron Classifier)

```python
from sklearn.neural_network import MLPClassifier

model2 = MLPClassifier(hidden_layer_sizes = (100, 50), 
                       solver = 'adam', 
                       activation = 'logistic', 
                       learning_rate = 'constant', 
                       batch_size = 'auto')
```

```python
model2.fit(x_train, y_train)
```

```python
y_pred = model2.predict(x_test)
y_pred
```

```python
accuracy2 = accuracy_score(y_test, y_pred)
accuracy2

print(f" Accuracy for MLP Classifier model is {accuracy2 * 100} %")
```

```python
cm2 = confusion_matrix(y_test, y_pred)
sns.heatmap(data = cm2, annot = True)
```

### Model 3 -- Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

model3 = RandomForestClassifier(n_estimators = 500, 
                                criterion = 'gini', 
                                max_depth = 10000, 
                                max_features = 'auto', 
                                bootstrap = True)

model3.fit(x_train, y_train)
```

```python
y_pred = model3.predict(x_test)
y_pred
```

```python
accuracy3 = accuracy_score (y_test, y_pred)
accuracy3

print(f"Random Forest Classifier Model's accuracy is {accuracy3 * 100} %  ")
```

```python
cm3 = confusion_matrix(y_test, y_pred)
cm3
```

```python
sns.heatmap(data = cm3, annot = True)
```
