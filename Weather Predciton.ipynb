```python
# So everthing Contains outlier values. Hence i'm doing the imputation process with median

imputer = SimpleImputer(strategy = 'median')
data[['WindSpeed9am','WindSpeed3pm','Humidity9am','Humidity3pm','Pressure9am','Pressure3pm']] = imputer.fit_transform(data[['WindSpeed9am','WindSpeed3pm','Humidity9am','Humidity3pm','Pressure9am','Pressure3pm']])
```

```python
data[['MinTemp','MaxTemp','Rainfall','WindGustSpeed','Temp9am','Temp3pm']] = imputer.fit_transform(data[['MinTemp','MaxTemp','Rainfall','WindGustSpeed','Temp9am','Temp3pm']])
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
# Encoding

from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
data['WindGustDir'] = encoder.fit_transform(data['WindGustDir'])
data['WindDir9am'] = encoder.fit_transform(data['WindDir9am'])
data['WindDir3pm'] = encoder.fit_transform(data['WindDir3pm'])
data['RainToday'] = encoder.fit_transform(data['RainToday'])
data['RainTomorrow'] = encoder.fit_transform(data['RainTomorrow'])

data['Location'] = encoder.fit_transform(data['Location'])
```

```python
data
```

```python
data.Date.dtype

data.Date = pd.to_datetime(data['Date'], infer_datetime_format = True)

data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month
data
```

```python
data = data.drop(['Date'], axis = 1)
data
```

```python
data.columns
```

```python
x = data[[ 'Location', 'MinTemp', 'MaxTemp', 'Rainfall', 'WindGustDir',
       'WindGustSpeed', 'WindDir9am', 'WindDir3pm', 'WindSpeed9am',
       'WindSpeed3pm', 'Humidity9am', 'Humidity3pm', 'Pressure9am',
       'Pressure3pm', 'Temp9am', 'Temp3pm', 'RainToday', 'RISK_MM', 'Year', 'Month']]

y = data['RainTomorrow']

y
```

```python
# Train Test Split

from sklearn.model_selection import train_test_split

x_train, x_test,y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 2)
```

```python
print('Train Independent shape: ',x_train.shape)
print('Train Dependent shape: ',y_train.shape)
print('Test Independent Shape: ', x_test.shape)
print('Test Dependent shape: ',y_test.shape)
```

```python
# feature Scaling 

from sklearn.preprocessing import StandardScaler

scale = StandardScaler()
x_train = scale.fit_transform(x_train)
x_test = scale.fit_transform(x_test)
```
